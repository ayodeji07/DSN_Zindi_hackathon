```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')
samplesubmission = pd.read_csv('SampleSubmission.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
drop_train = train.drop(["Any_Complain", "No_of_Kids_in_home", "Cmp2Accepted", "No_of_Teen_in_home", "Cmp4Accepted"], axis=1,) 
```

```python
drop_train.head()
```

```python
cat_feat = [feat for feat in drop_train.columns if train[feat].dtype == 'O']

cat_feat
```

```python
num_feat = [feat for feat in drop_train.columns if feat not in cat_feat]

num_feat
```

```python
rem = ['Response']
for i in rem:
    num_feat.remove(i)
num_feat
```

```python
updated_train = drop_train
updated_train['Disposable_Income'] = updated_train['Disposable_Income'].fillna(updated_train['Disposable_Income'].median())
```

```python
updated_test = test
updated_test['Disposable_Income']=updated_test['Disposable_Income'].fillna(updated_test['Disposable_Income'].median())

print(updated_test[num_feat].isnull().sum())
```

```python
scaler = StandardScaler()
cols = num_feat
X = scaler.fit_transform(updated_train[num_feat])
testing = scaler.transform(updated_test[num_feat])
X = pd.DataFrame(data=X, columns=cols)
testing = pd. DataFrame(data=updated_test, columns=cols)

X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, updated_train.Response, stratify=updated_train.Response, test_size=0.10, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rmmodel = RandomForestClassifier(n_estimators=100)
rmmodel.fit(X_train, y_train)
```

```python
pred1 = rmmodel.predict(X_test)
pred2 = rmmodel.predict(X_train)
```

```python
print('Training Set')
print('classifier f1_score {}'.format(f1_score(y_train,pred2)))
print(" ")
print('Testing Set')
print('classifier f1_score {}'.format(f1_score(y_test,pred1)))
```
